```python
boston = pd.read_csv("/kaggle/input/kaviya/HousingData (1).csv")
```

```python
boston.head(3)
```

```python
boston[['CRIM','ZN','INDUS','AGE','LSTAT']] = boston[['CRIM','ZN','INDUS','AGE','LSTAT']].fillna((boston[['CRIM','ZN','INDUS','AGE','LSTAT']].mean()))
```

```python
boston['CHAS']=boston['CHAS'].fillna(method='bfill')
# cols = ['CHAS']
# boston = boston.drop(cols, axis=1)
```

```python
X = boston.drop('MEDV', axis = 1)
Y = boston['MEDV']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)

from sklearn.ensemble import RandomForestRegressor


RF=RandomForestRegressor(n_estimators = 100, random_state = 10)
RF.fit(X_train,Y_train)
a = RF.score(X_test,Y_test)
print(a)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=123)

#subsample ratio
import xgboost as xgb
xg_reg = xgb.XGBRegressor(colsample_bytree = 0.7,learning_rate=0.1,max_depth=7,n_estimators=1000)
xg_reg.fit(X_train,Y_train)
b = xg_reg.score(X_test,Y_test)
print(b)
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=1)
from catboost import Pool, CatBoostRegressor
#When the number of iterations decreases, the learning rate needs to be increased.
c = CatBoostRegressor(iterations=40, depth=1, learning_rate=0.6)
c.fit(X_train, y_train)
c1 = c.score(X_test, y_test)
print(c1)
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size=0.2, random_state=42)
import lightgbm as lgb


gbm = lgb.LGBMRegressor(learning_rate=0.5,num_iterations = 40,n_estimators =100,max_depth= 5)
gbm.fit(X_train, y_train)
d = gbm.score(X_test, y_test)
print(d)
```

```python
Predictive_models = pd.DataFrame({
    'Model': [ 'Random Forest', 'Extreme Gradient Boosting', 'Cat Boost','Light GBM'],
    'Score': [a*100, b*100, c1*100, d*100]})
Predictive_models.sort_values(by ='Score',ascending = True)
```
